```python

def is_pos_int(value):
    """Check if the given value is a positive integer (including zero)."""
    # Check if the value is a positive integer (in this case, including zero)
    return isinstance(value, int) and value >= 0


def get_input(prompt="> "):
    """Get user input, strip leading/trailing whitespace, and convert to lowercase."""
    return input(prompt).strip().lower()


def show_census_data(data):
    """Display the census data in a formatted manner."""
    headers = "[home_cat][home_own][num_people][num_female][num_male][num_incomes][has_pet]"
    print(headers)
    for line in data:
        for item in line:
            print(f"{item:<12}", end="")
    print()  # New line after the records


def sort_census_data(data):
    """Sort the census data using bubble sort algorithm on."""
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            num_people_i = data[i][2]
            num_people_j = data[j][2]
            if num_people_i > num_people_j:
                data[i], data[j] = data[j], data[i]


def get_home_category():
    """Get a valid home category from the user, or 'END' to quit."""
    valid_home_cat = ["house", "apartment", "mobilehome"]
    while True:
        print("Enter home category (house, apartment, mobilehome) or 'END' to quit:")
        home_category = get_input()

        # Check for no input
        if home_category == "":
            print("No input provided. Please enter a home category.")
            continue

        # Check for the exit flag ("END")
        if home_category.lower() == "end":
            # Return the exit flag to indicate quitting
            return -1

        if home_category in valid_home_cat:
            return home_category
        else:
            print("Invalid home category. Please try again.")


def get_home_ownership():
    """Get a valid home ownership status from the user."""
    valid_home_own = ["owned", "mortgaged", "rented", "hotel", "occupied"]

    while True:
        print("Enter home ownership (owned, mortgaged, rented, hotel, occupied):")
        home_ownership = get_input()

        if home_ownership in valid_home_own:
            return home_ownership
        else:
            print("Invalid home ownership. Please try again.")


def get_num_people():
    """Get the number of people in the household from the user."""
    while True:
        print("Enter number of people in the household:")
        num_people = get_input()

        if not num_people.isdigit():
            print("Invalid input. Please enter a positive integer.")
            continue

        num_people = int(num_people)
        if not is_pos_int(num_people):
            print("Invalid input. Please enter a positive integer.")
            continue

        return num_people


def get_num_female(num_people):
    """Get the number of females in the household from the user."""
    while True:
        print("Enter number of females in the household:")
        num_females = get_input()

        if not num_females.isdigit():
            print("Invalid input. Please enter a positive integer.")
            continue

        num_females = int(num_females)
        if not is_pos_int(num_females):
            print("Invalid input. Please enter a positive integer.")
            continue

        if num_females > num_people:
            print(
                "Number of females cannot exceed total number of people. Please try again.")
            continue

        return num_females


def get_num_male(num_people, num_females):
    """Get the number of males in the household from the user."""
    while True:
        print("Enter number of males in the household:")
        num_males = get_input()

        if not num_males.isdigit():
            print("Invalid input. Please enter a positive integer.")
            continue

        num_males = int(num_males)
        if not is_pos_int(num_males):
            print("Invalid input. Please enter a positive integer.")
            continue

        if num_males > (num_people - num_females):
            print(
                "Number of males cannot exceed total number of people minus number of females. Please try again.")
            continue

        return num_males


def get_num_incomes(num_people):
    """Get the number of incomes in the household from the user."""
    while True:
        print("Enter number of incomes in the household:")
        num_incomes = get_input()

        if not num_incomes.isdigit():
            print("Invalid input. Please enter a positive integer.")
            continue

        num_incomes = int(num_incomes)
        if not is_pos_int(num_incomes):
            print("Invalid input. Please enter a positive integer.")
            continue

        if num_incomes > num_people:
            print(
                "Number of incomes cannot exceed total number of people. Please try again.")
            continue

        return num_incomes


def get_has_pet():
    """Get whether the household has a pet from the user."""
    while True:
        print("Does the household have a pet? (yes/no):")
        has_pet = get_input()

        if has_pet in ["yes", "no"]:
            return has_pet
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")


def get_sensus_data(data):
    """Collect census data from the user and store it in the provided data list."""
    # Get home category and check for quit flag
    record = []
    home_category = get_home_category()
    if home_category == -1:
        return home_category  # Return -1 to indicate quitting
    else:
        data.append(home_category)

    # Get remaining data fields
    home_ownership = get_home_ownership()
    record.append(home_ownership)
    num_people = get_num_people()
    record.append(num_people)
    num_females = get_num_female(num_people)
    record.append(num_females)
    num_males = get_num_male(num_people, num_females)
    record.append(num_males)
    num_incomes = get_num_incomes(num_people)
    record.append(num_incomes)
    has_pet = get_has_pet()
    record.append(has_pet)
    data.append(record)


def main():
    """Main function to run the census data collection program."""
    data = []

    while True:
        flag_to_quit = get_sensus_data(data)
        if flag_to_quit == -1:
            sort_census_data(data)
            show_census_data(data)
            break


# Run the program
if __name__ == '__main__':
    main()
```
