```python
def min_max(t):
    return min(t), max(t)

def mean(*args):
    return sum(args) / len(args)

def trimmed_mean(*args):
    low, high = min_max(args)
    trimmed = list(args)
    trimmed.remove(low)
    trimmed.remove(high)
    return mean(*trimmed)

test_data = (1, 2, 3, 10)
print(mean(*test_data))
print(trimmed_mean(*test_data))
```

```python
# INVERTING A DICTIONARY
def value_counts(string):
    counter = {}
    for letter in string:
        if letter not in counter:
            counter[letter] = 1
        else: 
            counter[letter] += 1
    return counter
    
d = value_counts('parrot')

def invert_dict(d):
    new = {}
    for key, value in d.items():
        if value not in new:
            new[value] = [key]
        else:
            new[value].append(key)
    return new

inverted = invert_dict(d)
print(d)
print(inverted) 
```

```python
# WORD DISTANCE
from doctest import run_docstring_examples

def run_doctests(func):
    run_docstring_examples(func, globals(), name=func.__name__)

def word_distance(word1, word2):
    """
    Computes the number of places where two word differ.

    >>> word_distance("hello", "hxllo")
    1
    >>> word_distance("ample", "apply")
    2
    >>> word_distance("kitten", "mutton")
    3
    """
    # Check the words are the same lenght
    if len(word1) != len(word2):
        return -1

    compare = list(zip(word1, word2))
    distance = 0
    for w1, w2 in compare:
        if w1 != w2:
            distance += 1
    return distance
    
    
run_doctests(word_distance)
```

```python
# METATHESIS
def is_anagram(word1, word2):
    return sorted(word1) == sorted(word2)  


# Copied from ChatGPT. I really had no idea how to do this. Would have taken me all day. 
def is_metathesis_pair(w1, w2):
    """Return true if w1 and w2 differ by exactly one swap"""
    if len(w1) != len(w2):
        return False

    diffs = []

    # Collect positions where they differ
    for i in range(len(w1)):
        if w1[i] != w2[i]:
            diffs.append(i)
            if len(diffs) > 2: 
                return False

    if len(diffs) != 2:
        return False

    i, j = diffs
    # Check if swapping the mismatched letters fixes the word
    return (w1[i] == w2[j] and w1[j] == w2[i])

def find_metathesis_pairs(words):
    """Return a list of metathesis pairs from the wordlist."""
    # Group the words that are anagrams
    anagram_groups = {}
    for word in words:
        key = ''.join(sorted(word))
        if key not in anagram_groups:
            anagram_groups[key] = []
        anagram_groups[key].append(word)

    # Check inside each anagram group for metathesis pairs
    pairs = []
    for group in anagram_groups.values():
        n = len(group)
        for i in range(n):
            for j in range(i + 1, n):
                if is_metathesis_pair(group[i], group[j]):
                    pairs.append((group[i], group[j]))

    return pairs

word_list = open("web2.txt").read().split()
pairs = find_metathesis_pairs(word_list)

for p,q in pairs:
    if len(p) > 12:
        print(p, q)
```
