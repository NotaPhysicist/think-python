```python

def is_good_email(email):
    """
    Returns True if email is a valid email address according to the specified rules.

    Args:
        email (string): input email address to check    

    Returns:
        bool: True if email is valid, False otherwise    
    """
    # Break email into local and domain
    subcomponents = email.split("@")

    # Chech that there are exactly two subcomponents
    if len(subcomponents) != 2:
        return False

    # Check that neither subcomponent is empty
    first_part, second_part = subcomponents
    if len(first_part) == 0 or len(second_part) == 0:
        return False

    # Check that both subcomponents contain only allowed characters
    if not is_Let_dig_und_stop(first_part):
        return False
    if not is_Let_dig_und_stop(second_part):
        return False

    # Check that both subcomponents do not start or end with a dot
    if first_part[0] == "." or first_part[-1] == ".":
        return False
    if second_part[0] == "." or second_part[-1] == ".":
        return False

    # Check that there are no consecutive dots in either subcomponent
    if ".." in first_part or ".." in second_part:
        return False

    return True


def is_Let_dig_und_stop(str):
    """
    Returns True if str contains only letters, digits, underscores, dots or stops.

    Args:
        str (string): input string to check

    Returns:
        bool: True if str contains only allowed characters, False otherwise
    """
    allowed = "abcdefghijklmnopqrstuvwxyz0123456789._"
    for char in str.lower():
        if char not in allowed:
            return False
    return True


print(is_good_email("kz.usq@gmail.com.au"))  # True
print(is_good_email("kz..usq@gmail.com"))    # False
print(is_good_email("kz.usq@gmail..com"))    # False
print(is_good_email("kz.usq@.gmail.com"))    # False
print(is_good_email("kz.usq@gmail.com."))    # False
```
